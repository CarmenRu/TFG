# PRUEBAS EN LAS BBDD

```python
import pandas as pd

enz = pd.read_csv("enzimas.csv")
```

```python
enz.head()
```

```python
total = len(enz["drugbank_id"].unique())
print(total)
```

```python
#Solo me quedo con las de tipo enzima
enzimas = enz[enz["type"]== "enzyme"]
enzimas_total = len(enzimas["drugbank_id"].unique())
print(enzimas_total)
```

```python
enzimas["action"].unique()
```

```python
#Cuantos hay de cada
diccionario = enzimas["gene_name"].value_counts().to_dict()
print(diccionario)
```

```python
# Hay mas de uno con los que me quedo en un mismo farmaco?
#Elimino los que no sean CYP2D6, CYP3A4, CYP3A5, CYP2D19
eliminados = enzimas[enzimas["gene_name"].isin(["CYP2D6", "CYP3A4", "CYP3A5", "CYP2D19"])]
reducido = len(eliminados["drugbank_id"].unique())
print(reducido)
#Me quedo solo con los duplicados incluyendo el primero
mas1 = eliminados[eliminados.duplicated(subset="drugbank_id", keep=False)]
```

```python
f_mas1 = len(mas1["drugbank_id"].unique())
print(f_mas1)
```

```python
print(f"Hay {total} fármacos en total. \nCon enzimas: {enzimas_total}. \nNos quedamos solo con: {reducido}. \nQue corresponden al {(reducido/enzimas_total)*100}% de todos los fármacos con enzimas.\n  Por último, fármacos metabolizados por mas de una enzima hay: {f_mas1} ")
```
